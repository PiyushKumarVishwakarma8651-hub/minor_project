##### Retail Store Inventory Analysis #####

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Day':[1,2,3,4,5,6,7],
    'Laptop_stock' :[50,45,42,38,15,12,10],
    'Phone_Stock': [30,27,25,22,20,28,15],
    'Min_Threshold':[20,20,20,20,20,20,20]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Shortage_Alert'] = df['Laptop_stock'] < df['Min_Threshold']
```

```python
print("--- Inventory Data ---")
print(df)
```

```python
laptop_variance = df['Laptop_stock'].var()
print("\nLaptop Stock Variance: {laptop_variance:.2f}")
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Day'],df['Laptop_stock'] , label = 'laptop Stock', marker='o',color='blue')
plt.plot(df['Day'],df['Phone_Stock'],label = 'Phone Stock', marker='s',color='red')
plt.axhline(y=20, color='red',linestyle='--',label='Minimum Threshold')
plt.title('Inventory level Over 7 days')
plt.xlabel('Days')
plt.ylabel('Units in Stock')
plt.legend()
plt.grid(True)
plt.show()
```
